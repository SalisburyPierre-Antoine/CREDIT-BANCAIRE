```python
# On vérifie maintenant qu'on a recodé toute les variables string en int que la corrélation peut se faire pour nos 106 colonnes
# On indique vouloir garder toutes les colonnes
pd.set_option('display.max_columns',None)
# On calcule la corrélation par paire des colonnes, à l'exclusion des valeurs NA/null.
coorMatrix = application.corr(method='pearson')
coorMatrix.shape
```

```python
# Nous avaons bien 122 colonnes donc aucune perte d'information, nous allons maintenant étudier les corrélations sur 
# notre variable cible : TARGET -> 0 : client en difficulté de paiement ; 1 : autres cas 
```

```python
# Calcul litérale 
# Le .abs nous permet de tout mettre en valeur absolue : si il y a un -1 (calcul négatif = anticorrélation) 
# ça donne aussi de l'information sur TARGET 
```

```python
# On cherche a calculer les corrélations entre la variable cible : TARGET 
# et les autres variables en prenant compte des valeurs absolues
# On regarde un peu comment sont réparties les valeurs 
coorMatrix['TARGET'].abs().describe()
```
