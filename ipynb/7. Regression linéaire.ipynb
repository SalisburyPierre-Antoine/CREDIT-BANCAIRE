```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier() # On vient de charger un model de classification
```

```python
y = rl['cible'] # On sélectionne la colonne cible
X = rl.drop('cible', axis=1) # On sélectionne toutes les colonnes sauf cible
```

```python
model.fit(X,y)
model.score(X,y)
```

```python
# On écrit une fonction pour prédir si notre client va avoir un défaut de paiement ou non.
# Secondary / secondary special = 0 Higher education = 1 Incomplete higher = 2 Lower secondary = 3 Academic degree = 4
# Laborers = 0 Core staff = 1 Accountants = 2 Managers = 3 Drivers = 4 Sales staff = 5 Cleaning staff = 6 Cooking 
# 7 Private service staff = 8 Medicine staff = 9 Security staff = 10 High skill tech staff = 11Waiters/barmen staff 
# 12 Low-skill Laborers = 13 Realty agents = 14 Secretaries = 15 IT staff = 16 HR staf = 17

def defaut(model,niveau_etude=2, genre=0, age=31, statut_matrimonial=0, categorie_pro=0  ):
    x = np.array([niveau_etude, genre, age, statut_matrimonial, categorie_pro]).reshape(1,5)
    print(model.predict(x))
    print(model.predict_proba(x))
```

```python
defaut(model) # On estime si je vais avoir un défaut de paiement mais ce n'est pas suffisant ..
              # .. on calcule la probabilité que j'ai un défaut de paiement
```
